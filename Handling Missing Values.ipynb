# Filling in missing values automatically

We can use the Panda's fillna() function to fill in missing values in a dataframe for us. One option we have is to specify what we want the NaN values to be replaced with. Here, I'm saying that I would like to replace all the NaN values with 0.

```python
# replace all NA's with 0
df_3 =original_df.fillna(0)
df_3
```

```python
# Strategy 3: Fill missing numerical values with median values
from sklearn.impute import SimpleImputer
df_4=original_df
num_cols = df_4.select_dtypes(include=['number']).columns
imputer = SimpleImputer(strategy='median')
df_4[num_cols] = imputer.fit_transform(df_4[num_cols])
print("\nNumerical missing values filled with median")
```

```python
df_4.describe()
```

```python
df_4.info()
```

```python
# Strategy 4: Fill missing categorical values with mode
df_5=original_df
cat_cols = df_5.select_dtypes(include=['object']).columns
for col in cat_cols:
    df_5[col].fillna(df_5[col].mode()[0], inplace=True)
print("\nCategorical missing values filled with mode")
```

```python
df_5.describe()
```

```python
# Strategy 5: Fill missing numerical values with mean
df_6=original_df
imputer_mean = SimpleImputer(strategy='mean')
df_6[num_cols] = imputer_mean.fit_transform(df_6[num_cols])
print("\nNumerical missing values filled with mean")
```

```python
df_6.describe()
```
