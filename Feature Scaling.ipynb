# Importing Libraries

```python
# modules we'll use
import pandas as pd
import numpy as np

# for Box-Cox Transformation
from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling

# plotting modules
import seaborn as sns
import matplotlib.pyplot as plt
```

# Scaling vs. Normalization: What's the difference?

The difference is that, in scaling, you're changing the range of your data while in normalization you're changing the shape of the distribution of your data.

# Scaling

This means that you're transforming your data so that it fits within a specific scale, like 0-100 or 0-1. You want to scale data when you're using methods based on measures of how far apart data points, like support vector machines, or SVM or k-nearest neighbors, or KNN. With these algorithms, a change of "1" in any numeric feature is given the same importance.

```python
# generate 1000 data points randomly drawn from an exponential distribution
original_data = np.random.exponential(size = 1000)

# mix-max scale the data between 0 and 1
scaled_data = minmax_scaling(original_data, columns = [0])

# plot both together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")
```

# Normalisation

Scaling just changes the range of your data. Normalization is a more radical transformation. The point of normalization is to change your observations so that they can be described as a normal distribution.

```python
# normalize the exponential data with boxcox
normalized_data = stats.boxcox(original_data)

# plot both together to compare
fig, ax=plt.subplots(1,2)
sns.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalized_data[0], ax=ax[1])
ax[1].set_title("Normalized data")
```

```python
# Drop unnecessary columns (Unnamed columns with mostly NaN values)
df_cleaned = df.drop(columns=['Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16'], errors='ignore')
```

```python
# Trim spaces from column names
df_cleaned.columns = df_cleaned.columns.str.strip()
```

```python
df_cleaned.head(10)
```

```python
# Convert numerical columns to appropriate types
numeric_cols = ['goal', 'pledged', 'backers', 'usd pledged']
for col in numeric_cols:
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

# Fill NaN values with 0 for numerical processing
df_cleaned[numeric_cols] = df_cleaned[numeric_cols].fillna(0)
```

```python
# Initialize scalers
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, MaxAbsScaler

minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
robust_scaler = RobustScaler()
maxabs_scaler = MaxAbsScaler()

# Apply scalers
df_scaled = df_cleaned.copy()
```

```python
df_scaled['goal_minmax'] = minmax_scaler.fit_transform(df_scaled[['goal']])
df_scaled['pledged_minmax'] = minmax_scaler.fit_transform(df_scaled[['pledged']])
df_scaled['backers_minmax'] = minmax_scaler.fit_transform(df_scaled[['backers']])
df_scaled['usd_pledged_minmax'] = minmax_scaler.fit_transform(df_scaled[['usd pledged']])
```

```python
df_scaled['goal_standard'] = standard_scaler.fit_transform(df_scaled[['goal']])
df_scaled['pledged_standard'] = standard_scaler.fit_transform(df_scaled[['pledged']])
df_scaled['backers_standard'] = standard_scaler.fit_transform(df_scaled[['backers']])
df_scaled['usd_pledged_standard'] = standard_scaler.fit_transform(df_scaled[['usd pledged']])
```

```python
df_scaled['goal_robust'] = robust_scaler.fit_transform(df_scaled[['goal']])
df_scaled['pledged_robust'] = robust_scaler.fit_transform(df_scaled[['pledged']])
df_scaled['backers_robust'] = robust_scaler.fit_transform(df_scaled[['backers']])
df_scaled['usd_pledged_robust'] = robust_scaler.fit_transform(df_scaled[['usd pledged']])
```

```python
df_scaled['goal_maxabs'] = maxabs_scaler.fit_transform(df_scaled[['goal']])
df_scaled['pledged_maxabs'] = maxabs_scaler.fit_transform(df_scaled[['pledged']])
df_scaled['backers_maxabs'] = maxabs_scaler.fit_transform(df_scaled[['backers']])
df_scaled['usd_pledged_maxabs'] = maxabs_scaler.fit_transform(df_scaled[['usd pledged']])
```

```python
# Visualization of scaling effects
scaling_methods = ['minmax', 'standard', 'robust', 'maxabs']
numerical_feature = 'goal'

plt.figure(figsize=(12, 6))
for i, method in enumerate(scaling_methods, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df_scaled[f'goal_{method}'], bins=50, kde=True)
    plt.title(f'Goal - {method.capitalize()} Scaling')
plt.tight_layout()
plt.show()
```
